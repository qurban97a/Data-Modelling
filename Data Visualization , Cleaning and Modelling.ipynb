# Data Cleaning and Modelling

```python
df["id"]
```

```python
df.drop("id",axis=1,inplace=True)
```

```python
car_name = pd.get_dummies(df['car_name'],drop_first=True)
yr_mfr= pd.get_dummies(df['yr_mfr'],drop_first=True)
fuel_type = pd.get_dummies(df['fuel_type'],drop_first=True)
city = pd.get_dummies(df['city'],drop_first=True)
body_type = pd.get_dummies(df['body_type'],drop_first=True)



df.drop(['car_name',"yr_mfr","fuel_type","city","body_type"],axis=1,inplace=True)
df = pd.concat([df,car_name,yr_mfr,fuel_type,city,body_type],axis=1)
```

```python
df.drop("model",axis=1,inplace=True)
```

```python
df["transmission"]=pd.factorize(df["transmission"])[0]
df["variant"]=pd.factorize(df["variant"])[0]
df["registered_city"]=pd.factorize(df["registered_city"])[0]
df["registered_state"]=pd.factorize(df["registered_state"])[0]
df["is_hot"]=pd.factorize(df["is_hot"])[0]
df["rto"]=pd.factorize(df["rto"])[0]
df["source"]=pd.factorize(df["source"])[0]
df["make"]=pd.factorize(df["make"])[0]
df["car_availability"]=pd.factorize(df["car_availability"])[0]
df["total_owners"]=pd.factorize(df["total_owners"])[0]
df["car_rating"]=pd.factorize(df["car_rating"])[0]
df["ad_created_on"]=pd.factorize(df["ad_created_on"])[0]
df["fitness_certificate"]=pd.factorize(df["fitness_certificate"])[0]
df["warranty_avail"]=pd.factorize(df["warranty_avail"])[0]
df["reserved"]=pd.factorize(df["reserved"])[0]
df["assured_buy"]=pd.factorize(df["assured_buy"])[0]
```

```python
price=df["sale_price"]
df.drop("sale_price",inplace=True,axis=1)
```

```python
df=pd.concat([df,price],axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
X=df.drop("sale_price",axis=1)
y=df["sale_price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train.shape
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
coeff_df = pd.DataFrame(reg.coef_, X.columns, columns=['Cofficent'])  
coeff_df
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.distplot(df['sale_price'])
import warnings
warnings.filterwarnings("ignore");
```

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
R2 = metrics.r2_score(y_test, y_pred)
print('R^2:' , R2)
n = X_test.shape[0] #sample size
p = X_test.shape[1] #number of predictors
print('Adjusted R^2 :' , 1-(1-R2)*(n-1)/(n-p-1))
print('Adjusted R^2 :'  , 1 - metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error:' , metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:' , metrics.mean_squared_error(y_test, y_pred)) 
print('Root Mean Squared Error:' , np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
